```python
# Read the CSV file
df = pd.read_csv('dog-data-set-borough-current-licenses.csv')
df.head()
```

```python
df.value_counts('AnimalGender')
```

```python
df_zipcode_level = df.pivot_table(index=['ZipCode','population','total_units', 'pct_white', 'pct_black', 'pct_hispanic', 'pct_asian', 'med_inc', 'med_age', 'owner_occupied', 'renter_occupied', 'poverty_total', 'poverty_below', 'edu_bachelor_up', 'Borough'],
               values='AnimalName',
               aggfunc='count')\
    .reset_index()\
    .rename(columns={'AnimalName':'dog_count'})\
    .sort_values(by='dog_count')

df_zipcode_level.head(25)
```

```python
# Filter for zip codes with more than 10 dogs
df_zipcodes_with_dogs = df_zipcode_level[df_zipcode_level['dog_count'] > 10]

# Export to CSV
df_zipcodes_with_dogs.to_csv('zipcodes_more_than_10_dogs.csv', index=False)
```

```python
# Read the CSV file
df_more_than_10_dogs = pd.read_csv('zipcodes_more_than_10_dogs.csv')
df_more_than_10_dogs.head()
```

```python
%%R 
# Here is a cell with a linear regression on mtcars to get you started
# mpg ~ wt + disp
# modify the formula and paste in subsequent cells to run different models
# you can use forward selection, backwards selection, or your own approach

df_regression <- df_more_than_10_dogs %>% 
    mutate(
        dog_rate = dog_count / population,
        pct_pov = poverty_below / population,
        pct_white = pct_white / population,
        pct_black = pct_black / population,
        pct_hispanic = pct_hispanic / population,
        pct_edu = edu_bachelor_up / population
    )

model <- lm("dog_rate ~  med_inc + pct_pov + pct_white + pct_edu", data=df_regression)
summary(model)
```

```python
%%R 


# residual analysis 
df_regression %>% 
    mutate(
        predict = predict(model),
        residual = resid(model),
        residual_z = scale(residual)
    ) %>% 
    arrange(residual) %>% 
    select(ZipCode, Borough, dog_rate, predict, residual, residual_z)
```

```python
%%R 

library(flexplot)

flexplot(dog_rate ~ med_inc, data=df_regression, method='lm') + 
    geom_label_repel(aes(label=ZipCode))
```
